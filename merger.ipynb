```python
import ROOT

# List of ROOT files with the histograms
root_files = ["bjet_pt_1.root", "bjet_pt_2.root", "bjet_pt_3.root", "bjet_pt_4.root", "bjet_pt_5.root"]
hist_name = "b_jet_pt_hist1"  # Histogram name in the ROOT files



# Initialize the merged histogram (from the first file)
f0 = ROOT.TFile.Open(root_files[0])
h0 = f0.Get("b_jet_pt_hist1")
hist_merged = h0.Clone("merged_hist")
hist_merged.Reset()       # Prevent double-counting
hist_merged.Add(h0)       # Now add the first histogram

# Loop through the rest
for fpath in root_files[1:]:
    f = ROOT.TFile.Open(fpath)
    h = f.Get("b_jet_pt_hist1")
    hist_merged.Add(h)

ROOT.gStyle.SetOptStat(0)
# Plot the merged histogram
c1 = ROOT.TCanvas()
c1.SetLogy()

hist_merged.Draw()
c1.Draw()
```

```python
import ROOT

# List of ROOT files with the histograms
root_files = ["met_1.root", "met_2.root", "met_3.root", "met_4.root", "met_5.root"]
hist_name = "hist1"  # Histogram name in the ROOT files

# Initialize the merged histogram (from the first file)
f0 = ROOT.TFile.Open(root_files[0])
hist_merged = f0.Get(hist_name).Clone("stats")
hist_merged.Reset()  # Start with an empty histogram

# Add the histograms from the other ROOT files
for root_file in root_files:
    f = ROOT.TFile.Open(root_file)
    hist = f.Get(hist_name)
    hist_merged.Add(hist)


ROOT.gStyle.SetOptStat(0)
# Plot the merged histogram
c2 = ROOT.TCanvas()
c2.SetLogy()

hist_merged.Draw()
c2.Draw()
```
